# **Tugas 2: Data Preparation**

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('/content/data_tugas 2.csv')
print('Shape dataset:', df.shape)
df.head(10)
```

# **1. Menampilkan Jumlah Data Hilang**

```python
# Ganti '?' dengan NaN
df.replace('?', np.nan, inplace=True)

# Tampilkan jumlah missing value setiap kolom
print(df.isnull().sum())

# Variabel dengan missing value terbanyak
print(f"\nVariabel dengan missing value terbanyak: '{df.isnull().sum().idxmax()}' ({df.isnull().sum().max()} missing values)")
```

Berdasarkan output di atas variabel normalized-losses yang memiliki data hilang paling banyak dengan jumlah 41 missing values dari total 205 baris data

# **2. Mengganti Missing Value pada Variabel Kategori**

```python
# Identifikasi missing value pada variabel kategorik
kategorik = ['make', 'bahan-bakar', 'jumlah-pintu', 'body-style']
print("Missing value sebelum diganti:")
print(df[kategorik].isnull().sum())

# Ganti missing value dengan 'unknown'
df[kategorik] = df[kategorik].fillna('unknown')

# Tabel ringkasan jumlah setiap kategori
for col in kategorik:
    print(f"\n--- {col} ---")
    print(df[col].value_counts())
```

Berdasarkan hasil identifikasi di atas, ditemukan missing value pada keempat variabel kategorik yaitu jumlah-pintu (18), bahan-bakar (13), body-style (12), dan make (4), di mana jumlah-pintu lah yang memiliki missing value terbanyak yaitu 18.

# **3. Mengganti Missing Value pada Variabel Numerik**

```python
# Variabel numerik bertipe float
numerik = ['length', 'width', 'height']
df[numerik] = df[numerik].astype(float)

# Hitung nilai rata-rata masing-masing variabel
print("Nilai rata-rata (mean) setiap variabel:")
for col in numerik:
    print(f"{col}: {df[col].mean():.4f}")

# Ganti missing value dengan nilai mean
for col in numerik:
    df[col] = df[col].fillna(df[col].mean())

# Cek missing value setelah imputasi
print("\nMissing value setelah imputasi:")
print(df[numerik].isnull().sum())
```

Berdasarkan output di atas sebelum imputasi, terdapat missing value pada variabel length (8), width (5), dan height (4). Semua data yang hilang tersebut kemudian diganti menggunakan nilai rata-rata (mean) masing-masing variabel, yaitu length sebesar 173.9025, width sebesar 65.9310, dan height sebesar 53.7373. Setelah proses imputasi selesai, seluruh missing value berhasil terisi sehingga jumlah data kosong pada ketiga variabel tersebut menjadi 0, yang artinya data sudah bersih

# **4. Min-Max Normalization**

```python
# Wheel-bases bertipe float
df['wheel-bases'] = df['wheel-bases'].astype(float)

# Min-Max Normalization: (x - min) / (max - min)
df['wheel-bases-normalized'] = (df['wheel-bases'] - df['wheel-bases'].min()) / \
                                (df['wheel-bases'].max() - df['wheel-bases'].min())

# Tampilkan hasil
print(df[['wheel-bases', 'wheel-bases-normalized']].head(10))
```

Kolom wheel-bases-normalized merupakan hasil Min-Max Normalization dari kolom wheel-bases, di mana setiap nilai diubah ke dalam skala 0 sampai 1. Semakin kecil nilai wheel-bases maka semakin kecil nilai normalized-nya, begitu pula sebaliknya. Contohnya, nilai wheel-bases 88.6 menghasilkan nilai normalized 0.058 (mendekati 0 karena termasuk nilai terkecil), sedangkan nilai 105.8 menghasilkan 0.559 yang berarti berada di tengah rentang data.

# **5. Z-Score Normalization**

```python
# Z-Score Normalization: (x - mean) / std
for col in ['length', 'width', 'height']:
    df[f'{col}-zscore'] = (df[col] - df[col].mean()) / df[col].std()

# Tampilkan hasil
print(df[['length', 'length-zscore', 'width', 'width-zscore', 'height', 'height-zscore']].head(10))
```

Kolom length-zscore, width-zscore, dan height-zscore merupakan hasil Z-Score Normalization dari masing-masing variabel. Nilai zscore menunjukkan seberapa jauh data tersebut dari nilai rata-ratanya. Nilai positif berarti data berada di atas rata-rata, nilai negatif berarti data berada di bawah rata-rata, dan nilai 0 berarti data tepat sama dengan rata-rata. Contohnya pada baris ke-7, nilai length sebesar 173.90 menghasilkan zscore 0.000 karena nilainya sama persis dengan mean, sedangkan nilai height 48.8 menghasilkan zscore -2.038 yang berarti nilainya jauh di bawah rata-rata.
